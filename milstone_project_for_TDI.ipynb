### How many different companies are represented in the data set?


```python
# first method
company_names = df['Company Name'].values
unique_company_names = set(company_names)

print('Number of companies are: ', len(unique_company_names))
```

```python
# second method
unique_co_names = []
co_cnt = 0
for co_name in company_names:
    if co_name not in unique_co_names:
        unique_co_names.append(co_name)
        co_cnt += 1
print('Number of companies are: ', co_cnt)
        
```

```python
# third method
print('Number of companies are: ', len(pd.unique(df['Company Name'])))
```

### What is the total number of jobs created for businesses in Queens?

```python
print('Number of job created in Queens: ', len(df[df['City'] == 'Queens'].dropna(subset = ['Job created'])))
```

### How many different unique email domains names are there in the data set?

```python
df_temp = df[['company email']].dropna()
df_temp[['email user', 'email domain']] = df_temp['company email'].str.split('@', expand = True)

print('Number of email domain names are: ', len(pd.unique(df_temp['email domain'])))
```

### Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
df['NTA_cnt'] = 1
df_grouped = df.groupby(['NTA'])
df_NTA5 = df_grouped.count()[['NTA_cnt']][df_grouped.count()['NTA_cnt']>5]
df_NTA5['Average_Total_Savings'] = df_grouped.mean()[['Total Savings']][df_grouped.count()['NTA_cnt']>5][['Total Savings']]
df_NTA5['Total_Job_Created'] = df_grouped.sum()[['Job created']][df_grouped.count()['NTA_cnt']>5]['Job created']
df_NTA5.reset_index(drop = False, inplace = True)
df_NTA5
```

## Plotting
Using the same data set and results that you were working with in the pandas action items section (Day 4), create a

scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.
histogram of the log of the average total savings.
line plot of the total jobs created for each month.

```python
import pandas as pd
from  matplotlib import pyplot as plt
df = pd.read_csv("Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_-_FY2020.csv")
pd.set_option('display.max_columns', 50)
df.head()
```

### scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.

```python
fig, ax = plt.subplots()
ax.scatter(df['Job created'], df['Total Savings'])
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Total savings')
ax.set_ylabel('Job created')
ax.set_title('Jobs created versus total saving')
ax.set_xlim(1,100)
ax.set_ylim(1,1e6)
plt.show()
```

### histogram of the log of the average total savings.


```python
fig, ax = plt.subplots()
ax.hist(df['Total Savings'].values, log = True, edgecolor = 'black')
ax.set_ylim(.1,1e3)
ax.set_xlabel('Average total savings')
```

### line plot of the total jobs created for each month.

```python
from matplotlib.ticker import MultipleLocator
df.index = pd.to_datetime(df['Effective Date'])
df_monthly = df.groupby(df.index.month).sum()[['Job created']]
df_monthly.reset_index(drop = False, inplace = True)
fig,ax = plt.subplots()
ax.plot(df_monthly['Effective Date'], df_monthly['Job created'])
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.set_xlim(1,12)
ax.set_xlabel('Month')
ax.set_ylabel('Job created')
plt.show()
```
